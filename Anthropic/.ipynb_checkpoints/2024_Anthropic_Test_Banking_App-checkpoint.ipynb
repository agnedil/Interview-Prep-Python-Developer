## Final Class for Levels 1 Through 3

```python
class BankingSystem:
    def __init__(self):
        # Mapping from account_id to its balance.
        self.accounts = {}
        # Mapping from account_id to the total outgoing amount
        # (money transferred out and withdrawn via pay).
        self.outgoing_transactions = {}
        # Counter for generating unique payment IDs.
        self.payment_count = 0
        # Mapping from cashback processing timestamp (as a string)
        # to a list of tuples (account_id, cashback_amount).
        self.cashback_events = {}

    def create_account(self, timestamp, account_id):
        """
        create_account(timestamp, account_id)
        Creates a new account with balance 0.
        Returns "True" if the account was created, or "False" if it already exists.
        """
        if account_id in self.accounts:
            return "False"
        self.accounts[account_id] = 0
        self.outgoing_transactions[account_id] = 0
        return "True"

    def deposit(self, timestamp, account_id, amount):
        """
        deposit(timestamp, account_id, amount)
        Deposits the given amount into the specified account.
        Returns a string representing the new balance, or an empty string if the account does not exist.
        """
        if account_id not in self.accounts:
            return ""
        self.accounts[account_id] += int(amount)
        return str(self.accounts[account_id])

    def transfer(self, timestamp, source_account_id, target_account_id, amount):
        """
        transfer(timestamp, source_account_id, target_account_id, amount)
        Transfers money from source_account_id to target_account_id.
        Returns the updated balance of the source account as a string if successful;
        otherwise, returns an empty string.
        Also updates the total outgoing for the source account.
        """
        amt = int(amount)
        if source_account_id not in self.accounts or target_account_id not in self.accounts:
            return ""
        if source_account_id == target_account_id or self.accounts[source_account_id] < amt:
            return ""
        self.accounts[source_account_id] -= amt
        self.accounts[target_account_id] += amt
        self.outgoing_transactions[source_account_id] += amt
        return str(self.accounts[source_account_id])

    def pay(self, timestamp, account_id, amount):
        """
        pay(timestamp, account_id, amount)
        Withdraws the specified amount from the account (if sufficient funds exist).
        Each successful withdrawal triggers a cashback of 2% (rounded down) that is scheduled to be
        credited 24 hours later (24*60*60*1000 = 86400000 milliseconds).
        Returns a unique payment identifier (e.g., "payment1") if the withdrawal is successful;
        otherwise, returns an empty string.
        Also, the withdrawn amount is added to the account's outgoing transactions.
        """
        if account_id not in self.accounts or self.accounts[account_id] < int(amount):
            return ""
        amt = int(amount)
        self.accounts[account_id] -= amt
        self.outgoing_transactions[account_id] += amt
        self.payment_count += 1
        payment_id = f"payment{self.payment_count}"
        # Calculate cashback: 2% of the withdrawn amount (rounded down).
        cashback_amount = amt // 50  # because 2% of amt = amt * 0.02 = amt/50.
        # Schedule cashback to be processed 24 hours later.
        cashback_time = str(int(timestamp) + 86400000)
        if cashback_time not in self.cashback_events:
            self.cashback_events[cashback_time] = []
        self.cashback_events[cashback_time].append((account_id, cashback_amount))
        return payment_id

    def process_cashbacks(self, timestamp):
        """
        process_cashbacks(timestamp)
        Processes any cashback events scheduled for the given timestamp.
        The timestamp is converted to a string; if any events are scheduled, they are processed
        by crediting the corresponding cashback amounts to the respective accounts.
        """
        ts_str = str(timestamp)
        if ts_str in self.cashback_events:
            for account_id, cashback_amount in self.cashback_events[ts_str]:
                if account_id in self.accounts:
                    self.accounts[account_id] += cashback_amount
            # Once processed, remove the events for that timestamp.
            del self.cashback_events[ts_str]

    def top_spenders(self, timestamp, n):
        """
        top_spenders(timestamp, n)
        Returns a string listing the top n accounts with the highest total outgoing transactions.
        The format is: "account_id1 (total_outgoing1), account_id2 (total_outgoing2), ..."
        Sorting is in descending order by total outgoing; in case of ties, sorted alphabetically.
        Before computing the ranking, cashback events scheduled for the given timestamp are processed.
        """
        n = int(n)
        self.process_cashbacks(timestamp)
        sorted_accounts = sorted(self.outgoing_transactions.items(), key=lambda item: (-item[1], item[0]))
        results = []
        for account_id, total_outgoing in sorted_accounts[:n]:
            results.append(f"{account_id} ({total_outgoing})")
        return ', '.join(results)
```

### Level 1

The banking system should support crating new acounts, depositing money into accounts and transfering money between accounts

You will receive a list of queries and the final output shouyld be an array of strings represinting the returned values of all queries. Every query will only call one operation.

All operations will have a timestamp parameters - a stringified timestamp in milliseconds. It is garanteed that all timestamps are unique and are in a range from 1 to 10^9. Operations will be given in order of strictly increasing timestamps

Initially, the baning system does not have any accounts, so implement operations to allow account creation, deposits and transfers between 2 different accounts.

def create_account(timestamp, account_id) should create a new account wil the given identifier it it doesn't already exist. Returns True if the account was successfully created or false if an account with account_id already exists

def deposit(timstamp, account_id, amount) should deposit the given amount of money to the specified account. Returns a string represenintg the balance of the account after the operation has been processed. If the spcified account doesn't exist, should retuen an empty string

def transfe(timestamp, source_account_id, target_account_id, amount) should transfer the given amount of money from account source_account_id to account target_account_id. Returns a string representing the balance of source_account_id if the transfer was succussful of am empty string otherwise, returnes an empty string if source_account_id or target_account_id doestn' exist, returns an empty strinf if source_account_id or target_account_id are the same, returns the same if source_account_id has insufficient funds to perform the transfer

Check for all possible edge cases

```python
class BankingSystem:
    def __init__(self):
        # Internal dictionary to store account balances
        self.accounts = {}

    def create_account(self, timestamp, account_id):
        """
        create_account(timestamp, account_id)
        Creates a new account with balance 0.
        Returns "True" if created successfully, or "False" if the account already exists.
        """
        if account_id in self.accounts:
            return "False"
        self.accounts[account_id] = 0
        return "True"

    def deposit(self, timestamp, account_id, amount):
        """
        deposit(timestamp, account_id, amount)
        Deposits the given amount into the specified account.
        Returns a string representing the updated balance, or an empty string if the account doesn't exist.
        """
        if account_id not in self.accounts:
            return ""
        self.accounts[account_id] += amount
        return str(self.accounts[account_id])

    def transfer(self, timestamp, source_account_id, target_account_id, amount):
        """
        transfer(timestamp, source_account_id, target_account_id, amount)
        Transfers money from source_account_id to target_account_id.
        Returns a string representing the source account's balance after the transfer if successful,
        or an empty string if the transfer fails (e.g. accounts don't exist, are the same, or insufficient funds).
        """
        if source_account_id not in self.accounts or target_account_id not in self.accounts:
            return ""
        if source_account_id == target_account_id or self.accounts[source_account_id] < amount:
            return ""
        self.accounts[source_account_id] -= amount
        self.accounts[target_account_id] += amount
        return str(self.accounts[source_account_id])

    def process_queries(self, queries):
        """
        Processes a list of queries.
        Each query is a tuple where the first element is the operation name.
        Returns a list of results for each query.
        """
        results = []
        for query in queries:
            op = query[0]
            if op == 'create_account':
                result = self.create_account(*query[1:])
                results.append(result)
            elif op == 'deposit':
                result = self.deposit(*query[1:])
                results.append(result)
            elif op == 'transfer':
                result = self.transfer(*query[1:])
                results.append(result)
            else:
                results.append("")
        return results
```

```python
queries = [
    ("create_account", "1", "123"),
    ("create_account", "2", "124"),
    ("deposit", "3", "123", 100),
    ("transfer", "4", "123", "124", 50),
    ("deposit", "5", "124", 150),
    ("transfer", "6", "124", "123", 200),
]

bank = BankingSystem()
result = bank.process_queries(queries)
print(result)
# Expected output: ["True", "True", "100", "50", "200", "0"]
```

### Level 2
Next, the same banking system must support ranking accounts based on outgoing transactions
def top_spenders(timestamp, n) should return the identifiers of the top n accounts with the highest outgoing transactions - the total amount of money either transferred out or paid / withdrawn (the PAY operation will be introduced later) - sorted in descending order or in case of a tie, sorted alphabetically by account_id in ascending order. The rest should be a string in the following format: "account_id1 (total_outgoing1), ... account_idN (total_outgoingN)".
If less than n accounts exist in the system, return all their identifiers (in the described format).
Cashback (an operation tht will be introduced later) should not be reflected in the calculations for total outgoing transactions

```python
class BankingSystem:
    def __init__(self):
        # Dictionary mapping account_id to its balance.
        self.accounts = {}
        # Dictionary mapping account_id to the total outgoing amount.
        self.outgoing_transactions = {}

    def create_account(self, timestamp, account_id):
        """
        create_account(timestamp, account_id)
        Creates a new account with a 0 balance.
        Returns "True" if the account was created or "False" if it already exists.
        """
        if account_id in self.accounts:
            return "False"
        self.accounts[account_id] = 0
        self.outgoing_transactions[account_id] = 0
        return "True"

    def deposit(self, timestamp, account_id, amount):
        """
        deposit(timestamp, account_id, amount)
        Deposits the given amount into the specified account.
        Returns the updated balance as a string or an empty string if the account does not exist.
        """
        if account_id not in self.accounts:
            return ""
        self.accounts[account_id] += int(amount)
        return str(self.accounts[account_id])

    def transfer(self, timestamp, source_account_id, target_account_id, amount):
        """
        transfer(timestamp, source_account_id, target_account_id, amount)
        Transfers money from source_account_id to target_account_id.
        Returns the updated balance of the source account as a string if successful,
        or an empty string if an error occurs (e.g. non-existent accounts, insufficient funds, or same account).
        Also updates the total outgoing amount for the source account.
        """
        amount = int(amount)
        if source_account_id not in self.accounts or target_account_id not in self.accounts:
            return ""
        if source_account_id == target_account_id or self.accounts[source_account_id] < amount:
            return ""
        self.accounts[source_account_id] -= amount
        self.accounts[target_account_id] += amount
        self.outgoing_transactions[source_account_id] += amount
        return str(self.accounts[source_account_id])

    def top_spenders(self, timestamp, n):
        """
        top_spenders(timestamp, n)
        Returns a string representing the top n accounts with the highest outgoing transactions.
        The accounts are sorted in descending order by total outgoing;
        in case of ties, sorted alphabetically by account_id.
        Format: "account_id1 (total_outgoing1), account_id2 (total_outgoing2), ..."
        If fewer than n accounts exist, returns all in the described format.
        """
        n = int(n)
        sorted_accounts = sorted(self.outgoing_transactions.items(), key=lambda item: (-item[1], item[0]))
        results = []
        for account_id, total_outgoing in sorted_accounts[:n]:
            results.append(f"{account_id} ({total_outgoing})")
        return ', '.join(results)

    def process_queries(self, queries):
        """
        Processes a list of queries and returns a list of string results.
        Supported operations: create_account, deposit, transfer, and top_spenders.
        """
        results = []
        for query in queries:
            op = query[0]
            if op == 'create_account':
                result = self.create_account(*query[1:])
                results.append(result)
            elif op == 'deposit':
                result = self.deposit(*query[1:])
                results.append(result)
            elif op == 'transfer':
                result = self.transfer(*query[1:])
                results.append(result)
            elif op == 'top_spenders':
                result = self.top_spenders(*query[1:])
                results.append(result)
            else:
                results.append("")
        return results
```

```python
queries = [
    ("create_account", "1", "123"),
    ("create_account", "2", "124"),
    ("deposit", "3", "123", "200"),
    ("transfer", "4", "123", "124", "150"),
    ("deposit", "5", "124", "300"),
    ("transfer", "6", "124", "123", "100"),
    ("top_spenders", "7", 2),
]
bank = BankingSystem()
result = bank.process_queries(queries)
print(result)
# Expected output:
# ["True", "True", "200", "50", "600", "200", "123 (150), 124 (100)"]
```

### Level 3
Now add `def pay(timestamp, account_id, amount)` which withdraws the given amount from the specified account. All withdraw transactions provide a 2% cashback - 2% of the withdrawn amount (rounded down to the nearest integer) will be refunded to the account 24 hours after the withdrawal. If the withdrawal is successful (the account holds the sufficient fuinds to withdraw th e given amount), return a string with a unique identifier of the payment transaction in this format:
"payment[ordinal number of withdraws from all accounts]" e.g. "payment1", "payment2" etc. Additional conditions:
return an empty string if account_id doens't exist or if account_id has insufficient funds
top_spenders should account for the total amount of money wihdrawn from accounts
the waiting period for cashback is 24 hours equal to `24*60*60*1000` milliseconds, so cashback will be processed at timestamp 86400000
When it's time to process cashback for a withdrawal, the amount must be refunded to the account before any other transactions are perfdormed at the releveant timestamnp

Also add `def name()`

```python
class BankingSystem:
    def __init__(self):
        # Mapping from account_id to its balance.
        self.accounts = {}
        # Mapping from account_id to the total outgoing amount
        # (money transferred out and withdrawn via pay).
        self.outgoing_transactions = {}
        # Counter for generating unique payment IDs.
        self.payment_count = 0
        # Mapping from cashback processing timestamp (as a string)
        # to a list of tuples (account_id, cashback_amount).
        self.cashback_events = {}

    def create_account(self, timestamp, account_id):
        """
        create_account(timestamp, account_id)
        Creates a new account with balance 0.
        Returns "True" if the account was created, or "False" if it already exists.
        """
        if account_id in self.accounts:
            return "False"
        self.accounts[account_id] = 0
        self.outgoing_transactions[account_id] = 0
        return "True"

    def deposit(self, timestamp, account_id, amount):
        """
        deposit(timestamp, account_id, amount)
        Deposits the given amount into the specified account.
        Returns a string representing the new balance, or an empty string if the account does not exist.
        """
        if account_id not in self.accounts:
            return ""
        self.accounts[account_id] += int(amount)
        return str(self.accounts[account_id])

    def transfer(self, timestamp, source_account_id, target_account_id, amount):
        """
        transfer(timestamp, source_account_id, target_account_id, amount)
        Transfers money from source_account_id to target_account_id.
        Returns the updated balance of the source account as a string if successful;
        otherwise, returns an empty string.
        Also updates the total outgoing for the source account.
        """
        amt = int(amount)
        if source_account_id not in self.accounts or target_account_id not in self.accounts:
            return ""
        if source_account_id == target_account_id or self.accounts[source_account_id] < amt:
            return ""
        self.accounts[source_account_id] -= amt
        self.accounts[target_account_id] += amt
        self.outgoing_transactions[source_account_id] += amt
        return str(self.accounts[source_account_id])

    def pay(self, timestamp, account_id, amount):
        """
        pay(timestamp, account_id, amount)
        Withdraws the specified amount from the account (if sufficient funds exist).
        Each successful withdrawal triggers a cashback of 2% (rounded down) that is scheduled to be
        credited 24 hours later (24*60*60*1000 = 86400000 milliseconds).
        Returns a unique payment identifier (e.g., "payment1") if the withdrawal is successful;
        otherwise, returns an empty string.
        Also, the withdrawn amount is added to the account's outgoing transactions.
        """
        if account_id not in self.accounts or self.accounts[account_id] < int(amount):
            return ""
        amt = int(amount)
        self.accounts[account_id] -= amt
        self.outgoing_transactions[account_id] += amt
        self.payment_count += 1
        payment_id = f"payment{self.payment_count}"
        # Calculate cashback: 2% of the withdrawn amount (rounded down).
        cashback_amount = amt // 50  # because 2% of amt = amt * 0.02 = amt/50.
        # Schedule cashback to be processed 24 hours later.
        cashback_time = str(int(timestamp) + 86400000)
        if cashback_time not in self.cashback_events:
            self.cashback_events[cashback_time] = []
        self.cashback_events[cashback_time].append((account_id, cashback_amount))
        return payment_id

    def process_cashbacks(self, timestamp):
        """
        process_cashbacks(timestamp)
        Processes any cashback events scheduled for the given timestamp.
        The timestamp is converted to a string; if any events are scheduled, they are processed
        by crediting the corresponding cashback amounts to the respective accounts.
        """
        ts_str = str(timestamp)
        if ts_str in self.cashback_events:
            for account_id, cashback_amount in self.cashback_events[ts_str]:
                if account_id in self.accounts:
                    self.accounts[account_id] += cashback_amount
            # Once processed, remove the events for that timestamp.
            del self.cashback_events[ts_str]

    def top_spenders(self, timestamp, n):
        """
        top_spenders(timestamp, n)
        Returns a string listing the top n accounts with the highest total outgoing transactions.
        The format is: "account_id1 (total_outgoing1), account_id2 (total_outgoing2), ..."
        Sorting is in descending order by total outgoing; in case of ties, sorted alphabetically.
        Before computing the ranking, cashback events scheduled for the given timestamp are processed.
        """
        n = int(n)
        self.process_cashbacks(timestamp)
        sorted_accounts = sorted(self.outgoing_transactions.items(), key=lambda item: (-item[1], item[0]))
        results = []
        for account_id, total_outgoing in sorted_accounts[:n]:
            results.append(f"{account_id} ({total_outgoing})")
        return ', '.join(results)
```

```python
queries = [
    ("create_account", "1000", "123"),
    ("create_account", "2000", "124"),
    ("deposit", "3000", "123", "500"),
    ("pay", "4000", "123", "200"),
    ("transfer", "5000", "123", "124", "100"),
    # A deposit in the far future to trigger cashback processing.
    ("deposit", "1500000000000", "124", "300"),
    # 24 hours after the payment should trigger cashback processing.
    ("top_spenders", "1500000864000", 2)
]
bank = BankingSystem()
results = []
for query in queries:
    # The timestamp is always the second element.
    timestamp = query[1]
    op = query[0]
    if op == 'create_account':
        result = bank.create_account(timestamp, *query[2:])
        results.append(result)
    elif op == 'deposit':
        result = bank.deposit(timestamp, *query[2:])
        results.append(result)
    elif op == 'transfer':
        result = bank.transfer(timestamp, *query[2:])
        results.append(result)
    elif op == 'pay':
        result = bank.pay(timestamp, *query[2:])
        results.append(result)
    elif op == 'top_spenders':
        result = bank.top_spenders(timestamp, *query[2:])
        results.append(result)
    else:
        results.append("")
print(results)
# Example expected output (actual numbers may vary depending on event timing):
# [
#   "True", "True", "500", "payment1", "200", "300", "123 (300), 124 (0)"
# ]
```
