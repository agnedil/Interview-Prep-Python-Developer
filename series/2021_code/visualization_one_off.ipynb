```python
y1 = [i[0] for i in metrics]
y2 = [i[1] for i in metrics]
print('MSE scores:', y1)
print('Log cosh scores:', y2)
x = range(0,len(y1))
xtick_labels = [i[2] for i in metrics]

plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
plt.plot(y1)
plt.xticks(x, xtick_labels)
plt.title('HP Gridsearch: Number of Units (TimeSeries)')
plt.ylabel('MSE')
plt.xlabel('Epochs')

plt.subplot(1, 2, 2)
plt.plot(y2)
plt.xticks(x, xtick_labels)
plt.title('HP Gridsearch: Number of Units (TimeSeries)')
plt.ylabel('Logcosh')
plt.xlabel('Epochs')

plt.tight_layout()
plt.show()
```
